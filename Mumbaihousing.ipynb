```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/house_price_mumbai.csv')
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['Location'].unique()
```

```python
df['BHK'].unique()
```

```python
def fun(x):
  if len(x)!=1:
    s = str(x).split(" ")[0]
    return s
  else:
    return x
```

```python
df['BHK'] = df['BHK'].map(lambda x: fun(x))
```

```python
df['BHK'].unique()
```

```python
i = df[df['BHK']=='R'].index
df.drop(i,axis=0,inplace=True)
```

```python
df.shape
```

```python
df['BHK'].unique()
```

```python
i = df[df['project'].isnull()].index
df.drop(i,axis=0,inplace=True)
```

```python
df.shape
```

```python
df['Location'] = df['Location'].map(lambda x: str(x).title())
```

```python
df['Total sqft'].unique()
```

```python
df['project'] = df['project'].map(lambda x: str(x).strip())
```

```python
df['project'].unique()
```

```python
df = df.reset_index()
```

```python
df.drop('index',inplace=True,axis=1)
```

```python
df.shape
```

```python
df['project'] = df['project'].map(lambda x: str(x).strip())
```

```python
for i,v in enumerate(df['Location']):
  df['project'][i] = df['project'][i].split(v)[0]
```

```python
df
```

```python
def fun(x):
  if 'Cr' in x or 'cr' in x:
    s=str(x).split(" ")[0]
    s1 = str(int(float(s)*100))
    return s1
  else:
    s = str(x).split(" ")[0]
    return s
```

```python
df['price']=df['price'].map(lambda x: fun(x))
```

```python
df['price']
```

```python
df['price_sqft'] = df['price_sqft'].map(lambda x: str(x).strip())
```

```python
df
```

```python
def fun(x):
  if "," in str(x):
    s = ''.join(str(x).split(","))
    return s
  else:
    return x
```

```python
df['price_sqft'] = df['price_sqft'].map(lambda x: fun(x))
```

```python
df
```

```python
df['BHK'] = df['BHK'].astype('int')
df['Total sqft'] = df['Total sqft'].astype('int')
df['price_sqft'] = df['price_sqft'].astype('int')
df['price'] = df['price'].astype('float')
```

```python
plt.figure(figsize=(12,10))
most_cat = df['Location'].value_counts()[:40]
sns.barplot(x=most_cat, y=most_cat.index)
plt.xticks(size=12)
plt.xlabel("Frequency",size=22,c="g")
plt.ylabel("Location",size=22,c="g")
plt.title("Total No of House Availability",size=28,c="r")
plt.show()
```

```python
plt.figure(figsize=(8,8))
labels =df['BHK'].value_counts(sort = True).index
sizes = df['BHK'].value_counts(sort = True)
plt.pie(sizes, labels=labels,autopct='%1.1f%%', shadow=False, startangle=270)
plt.title('Total % of BHK',size = 22)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Total sqft',y='price',data=df,hue='BHK',s=90)
plt.xlabel("Total Sqfeet Area",size=15,c="g")
plt.ylabel("Avg Price (Lakh)",size=15,c="g")
plt.show()
```

```python
plt.figure(figsize=(16,8))
df.groupby("Location")['price_sqft'].mean().sort_values(ascending=False)[:40].plot(kind="bar")
plt.xlabel("Location",size=15,c="g")
plt.ylabel("Avg Price (Lakh)",size=15,c="g")
plt.title("Avg Price for Each Location",size=28,c="k")
plt.show()
```

```python
plt.figure(figsize=(10,7))
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True,linewidths=.5,fmt='.1f')
plt.title("Correlation Graph",c="r",size=25)
plt.show()
```
